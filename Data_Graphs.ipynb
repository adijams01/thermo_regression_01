```python
# Import required libraries
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
project_01 = pd.read_csv("https://raw.githubusercontent.com/adijams01/thermo_regression_01/main/regression_csv_01.csv")
```

```python
project_01.head()
```

```python
x=project_01.drop("temperature",axis=1)
x1=project_01["latitude"]
x2=project_01["altitude"]
x3=project_01["precipitation"]
x4=project_01["humidity"]
x5=project_01["windspeed"]
y=project_01["temperature"]
```

```python
x1
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x1,y,c="y",label='Latitude')
plt.xlabel("Latitude")
plt.ylabel("Temperature")
plt.legend();
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x2,y,c="r",label='Altitude')
plt.xlabel("Altitude")
plt.ylabel("Temperature")
plt.legend();
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x3,y,c="b",label='Precipitation')
plt.xlabel("precipitation")
plt.ylabel("Temperature")
plt.legend();
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x4,y,c="g",label='Humidity')
plt.xlabel("Humidity")
plt.ylabel("Temperature")
plt.legend();
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x5,y,c="b",label='Windspeed')
plt.xlabel("Windspeed")
plt.ylabel("Temperature")
plt.legend();
```
